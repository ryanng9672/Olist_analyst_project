```python
df_payments=pd.read_csv(r"C:\Users\genhk\Downloads\Dataset\olist_order_payments_dataset.csv")
#df_reviews=pd.read_csv(r"C:\Users\genhk\Desktop\JDE9\carousell\Dataset\olist_order_reviews_dataset.csv")
df_customers=pd.read_csv(r"C:\Users\genhk\Downloads\Dataset\olist_customers_dataset.csv")
df_items=pd.read_csv(r"C:\Users\genhk\Downloads\Dataset\olist_order_items_dataset.csv")
df_products=pd.read_csv(r"C:\Users\genhk\Downloads\Dataset\olist_products_dataset.csv")


#group the orders,order_payments,order_items,call df_1
df_1=df_orders.merge(df_payments,on='order_id').merge(df_items,on='order_id')

#group the df_1 and products,call df_2
df_2=pd.merge(df_1,df_products)

#group the df_3 and customers，call df_3
df_3=pd.merge(df_customers,df_2)
df_3.head(10)
```

```python
#def missing data
def missing_info(data,num):


    null_data = clean_df.isnull().sum().sort_values(ascending=False)
    percent = clean_df.isnull().sum()/data.isnull().count()
    missing_data = pd.concat([null_data,percent.apply(lambda x: format(x, '.2%'))],axis=1,keys=['T_missing','missing %'])
    print(missing_data.head(num))

missing_info(clean_df,30)
```

```python
import matplotlib.pyplot as plt
import pandas as pd

na_df_valid_orders = na_df[na_df['order_status'] == 'delivered']


monthly_orders = na_df_valid_orders.groupby(['year', 'month']).size()


monthly_coupon_usage = na_df_valid_orders[na_df_valid_orders['payment_type'] == 'voucher'].groupby(['year', 'month']).size()


monthly_data = pd.DataFrame({
    'total_orders': monthly_orders,
    'coupon_usage': monthly_coupon_usage
}).reset_index()


monthly_data = monthly_data.fillna(0)


monthly_data['year'] = monthly_data['year'].astype(int)
monthly_data['month'] = monthly_data['month'].astype(int)


monthly_data['coupon_usage_rate'] = (monthly_data['coupon_usage'] / monthly_data['total_orders']) * 100


fig, ax1 = plt.subplots(figsize=(12, 6))


monthly_data['total_orders'].plot(kind='bar', color='skyblue', ax=ax1)


ax1.set_xlabel('Year-Month')
ax1.set_ylabel('Total Orders', color='skyblue')
ax1.tick_params(axis='y', labelcolor='skyblue')
ax1.set_xticklabels([f"{year}-{month:02d}" for year, month in zip(monthly_data['year'], monthly_data['month'])], rotation=45)


ax2 = ax1.twinx()
monthly_data['coupon_usage_rate'].plot(kind='line', marker='o', color='red', ax=ax2)


for i, txt in enumerate(monthly_data['coupon_usage_rate']):
    ax2.annotate(f"{txt:.2f}%", 
                 (i, txt),
                 textcoords="offset points", 
                 xytext=(0,10), 
                 ha='center')
    
ax2.set_ylabel('Coupon Usage Rate (%)', color='red')
ax2.tick_params(axis='y', labelcolor='red')
ax1.legend(['Total Orders'], loc='upper left')
ax2.legend(['Coupon Usage Rate (%)'], loc='upper right')
plt.title('Monthly Total Orders and Coupon Usage Rate')
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_extraction.text import TfidfVectorizer
from scipy.sparse import hstack


merged_df['order_purchase_hour'] = merged_df['order_purchase_timestamp'].dt.hour
merged_df['order_purchase_dayofweek'] = merged_df['order_purchase_timestamp'].dt.dayofweek
numeric_features = [
    'price',
    'freight_value',
    'payment_value',
    'payment_installments',
    'payment_sequential',
    'order_purchase_hour',
    'order_purchase_dayofweek'
]


X_numeric = merged_df[numeric_features].fillna(0)  
X_text_message = merged_df['review_comment_message'].fillna('')  
X_text_title = merged_df['review_comment_title'].fillna('')  
y = merged_df['review_score']  

X_train, X_test, y_train, y_test = train_test_split(merged_df, y, test_size=0.3, random_state=101)

# Now separate the numeric and text features from the TRAIN set
X_train_numeric = X_train[numeric_features].fillna(0)
X_train_text_message = X_train['review_comment_message'].fillna('')
X_train_text_title = X_train['review_comment_title'].fillna('')

# And also separate the text features from the TEST set
X_test_text_message = X_test['review_comment_message'].fillna('')
X_test_text_title = X_test['review_comment_title'].fillna('')

# Vectorize the text features for TRAIN set
tfidf_vectorizer_message = TfidfVectorizer(max_features=100, stop_words='english')
tfidf_vectorizer_title = TfidfVectorizer(max_features=50, stop_words='english')

tfidf_matrix_message_train = tfidf_vectorizer_message.fit_transform(X_train_text_message)
tfidf_matrix_title_train = tfidf_vectorizer_title.fit_transform(X_train_text_title)

# Vectorize the text features for TEST set using the fitted vectorizer
tfidf_matrix_message_test = tfidf_vectorizer_message.transform(X_test_text_message)
tfidf_matrix_title_test = tfidf_vectorizer_title.transform(X_test_text_title)

# Stack numeric and vectorized text features for the TRAIN set
X_train_combined = hstack([X_train_numeric, tfidf_matrix_message_train, tfidf_matrix_title_train])

# Stack numeric and vectorized text features for the TEST set
X_test_numeric = X_test[numeric_features].fillna(0)
X_test_combined = hstack([X_test_numeric, tfidf_matrix_message_test, tfidf_matrix_title_test])


rf = RandomForestClassifier(n_estimators=100, random_state=101)
rf.fit(X_train_combined, y_train)
predictions = rf.predict(X_test_combined)
accuracy = accuracy_score(y_test, predictions)
conf_matrix = confusion_matrix(y_test, predictions)
class_report = classification_report(y_test, predictions)


print(f"Accuracy: {accuracy}")
print("Confusion Matrix:")
print(conf_matrix)
print("Classification Report:")
print(class_report)

```
